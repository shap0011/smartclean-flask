#### Normalize and standardize data if necessary

```python
df.info()
df.head()
```

```python
# convert 'Transaction Date' column to proper format

# identify invalid date values
# import the regular expression module
import re

# define a regex pattern for valid date formats
date_pattern =  r"(\d{4}-\d{2}-\d{2})|(\d{2}/\d{2}/\d{4})|(\d{2}-\d{2}-\d{4})"

# find rows where the 'Transaction Date' column's value does not match a valid date format
invalid_values = df[~df['Transaction Date'].astype(str).str.match(date_pattern, na=False)]['Transaction Date'].unique()

# display the list of invalid values
print(f"List of invalid values in 'Transaction Date column': {invalid_values}")
```

```python
# replace all invalid values with NaN
df['Transaction Date'] = df['Transaction Date'].replace(invalid_values, pd.NA, regex=True)
```

```python
# convert to datetime, forcing invalid dates to NaT
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors="coerce") # not displaying NaT correctly, shows as NaN
```

```python
# consirm the 'Transaction Date' column data type
print(df['Transaction Date'].dtype)
```

```python
# replace NaN with a default date '2000-01-01'
# df['Transaction Date'].fillna(pd.to_datetime('2000-01-01'))
df['Transaction Date'] = df['Transaction Date'].fillna(pd.to_datetime('2000-01-01'))
```

```python
# standardize text columns
list_attr_to_stand = ['Item','Payment Method','Location']
for a in list_attr_to_stand:
    df[a] = df[a].str.strip().str.lower()
```

```python
# get summary statistics
df.describe(include="all")
```

```python
# convert data type of 'Quantity','Price Per Unit','Total Spent' columns to numeric format
# remove symbols and spaces before conversion
list_attr_to_numeric = ['Quantity','Price Per Unit','Total Spent']
for a in list_attr_to_numeric:
    df[a] = df[a].astype(str).str.replace(r"[^\d.]", "", regex=True)  # only numbers & decimal points
    df[a] = pd.to_numeric(df[a], errors="coerce")  # convert to numeric
```

```python
# replace NaN with the column median
for a in list_attr_to_numeric:
    print(f"{a}: {df[a].unique()}")
```

```python
# replace missing values NaN with the column median
for a in list_attr_to_numeric:
    df[a] = df[a].fillna(df[a].median())
```

### Convert Categorical data into numerical formats

```python
df.info()
```

```python
df.info()
```

```python
categorical_columns = ['Item','Payment Method','Location']
for c in categorical_columns:
    print(f"{c}: {df[c].unique()}")    
```

```python
for c in categorical_columns:
    unknown_count = (df[c] == "unknown").sum()
    print(f"{c} 'unknown': {unknown_count}")
    error_count = (df[c] == "error").sum()
    print(f"{c} 'error': {unknown_count}")
    total_count = unknown_count + error_count
    print(f"{c} total: {total_count}")
```

```python
# count percentages of inconsistancies 
print(585/9059)
print(555/9059)
print(632/9059)
```

```python
# the data quality is more important than keeping records
# drop rows with 'error' and 'unknown' values
df = df[~df[categorical_columns].isin(["unknown", "error"]).any(axis=1)]
print(f"New dataset size: {df.shape}")
```

```python
for c in categorical_columns:
    print(f"{c}: {df[c].unique()}") 
```

```python
# convert categorical data into numerical formats
df = pd.get_dummies(df, columns=categorical_columns)
```

```python
# get summary statistics
df.describe(include="all")
```

```python
# import library for visualizing
import matplotlib.pyplot as plt

# select only the encoded columns (binary 0/1 columns)
encod_columns = df.iloc[:, 4:]  # first 4 columns are numerical

# calculate the percentage of '1' (True) values in each column
encod_percentages = encod_columns.mean() * 100

# Plot a bar chart of category distributions
plt.figure(figsize=(12, 6))
encod_percentages.sort_values().plot(kind="bar", color="skyblue", edgecolor="black")

# formatting
plt.title("Distribution of Encoded Categories (% of '1' values)", fontsize=14)
plt.ylabel("Percentage of '1' values", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# show plot
plt.show()
```

### Handle outliers using statistical techniques.

```python
# detect outliers
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Quantity"])
plt.title("Box Plot of Quantity")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Price Per Unit"])
plt.title("Box Plot of Price Per Unit")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Total Spent"])
plt.title("Box Plot of Total Spent")
plt.show()
```

```python
# Calculate Q1 and Q3
Q1 = df["Total Spent"].quantile(0.25)
Q3 = df["Total Spent"].quantile(0.75)
IQR = Q3 - Q1

# Define outlier range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers_iqr = df[(df["Total Spent"] < lower_bound) | (df["Total Spent"] > upper_bound)]

print(f"Total outliers detected in 'Total Spent': {outliers_iqr.shape[0]}")
print(f"Outlier range: {lower_bound:.2f} to {upper_bound:.2f}")
print(f"Outlier values: {outliers_iqr['Total Spent'].unique()}")
```

```python
# drop outliers from 'Total Spent'
df = df[(df["Total Spent"] >= lower_bound) & (df["Total Spent"] <= upper_bound)]

print(f"Outliers removed. New dataset size: {df.shape}")
```

```python
# box plot to verify outliers are removed
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Total Spent"])
plt.title("Box Plot of Total Spent After Removing Outliers")
plt.show()
```
